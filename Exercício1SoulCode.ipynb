### **Exercício 01**

Você trabalha em um laboratório químico e toda vez que você realiza um experimento as temperaturas
estão em Fahrenheit ou Celsius. Crie um algoritmo que realize a conversão para Kelvin para facilitar sua vida de quem lê os manuais.
Opera Instantâneo_2024-04-02_201815_www.infoescola.com.png





```python
# Conversão (Fonte de conversões: https://www.infoescola.com/fisica/conversao-de-escalas-termometricas/)

# entrada de codigos
unidade = input("Digite a unidade de temperatura que deseja converter (F para Fahrenheit ou C para Celsius): ")
temp = float(input("Digite a temperatura: "))


# Conversão
if unidade == 'f':
    temperatura_kelvin = (temp - 32) * 5/9 + 273
    print('A temperatura em Kelvin é:', temperatura_kelvin)
elif unidade == 'c':
    temperatura_kelvin = temp + 273
    print('A temperatura em Kelvin é:', temperatura_kelvin)
else:
    print('Unidade de temperatura inválida, tente novamente: ')
```

## **Exercício 03**
Um aplicativo de compras online deseja implementar uma função de decisão de compra para seus usuários. Ao realizar uma compra, os usuários precisam saber se possuem saldo suficiente em sua conta bancária para comprar um determinado produto. Se o saldo disponível na conta bancária do usuário for maior ou igual ao preço do produto, ele poderá efetuar a compra. Caso contrário, o usuário receberá uma mensagem informando que não possui saldo suficiente para realizar a compra. O aplicativo precisa implementar essa funcionalidade para garantir uma experiência de compra tranquila e transparente para seus usuários.

```python
produto = float(input('Digite o valor do produto: '))
print(f'O produto é no valor de: {produto:.2f}')
saldo_em_conta = float(input('Digite o valor total disponivel em conta: '))
print(f'Seu saldo em conta é {saldo_em_conta:.2f}')

if saldo_em_conta >= produto:
    print('Compra Aprovada.')
else:
    print('Você não possui saldo suficiente. :( ')
```

## **Exercício 04**

Uma assistente do banco precisa de um programa que verifica se um cliente é elegível para um empréstimo com base na sua idade (entre 18 e 65 anos), renda mensal mínima de R$ 1.500,00 e histórico de crédito sem registros de inadimplência nos últimos 12 meses. O programa irá coletar as informações do cliente e, caso ele atenda a todos os critérios, será considerado elegível para o empréstimo; caso contrário, será informado que o cliente não é elegível. Essa verificação é essencial para garantir que apenas clientes adequados sejam aprovados para empréstimos, minimizando riscos para o banco e para o cliente.

```python
# Entrada de dados...
idade = int(input(' Idade: '))
if idade >= 18:
    print('Você é maior de idade!')
elif idade >= 66:
    print('Idade não permitida para empréstimo')
else:
    print('Você é menor de idade!')

saldo = float(input(' Saldo ativo: R$ '))
if saldo >= 1.500:
    print('Saldo suficiente!')

else:
    print('Saldo insuficiente!')
credito = str(input(' Contem inadimplências? '))
if credito == 'não':
    print('Histórico de crédito positivo!')

else:
    print('Histórico de crédito negativado!')

if idade >= 18 and saldo >= 1.500 and credito == 'não':
    print('Empréstimo aprovado!')

else:
    print('Empréstimo recusado!')
```

## **Exercício 05**

Crie um programa que simula a venda de um produto. O programa deve:
Coletar informações do usuário: nome do cliente, nome do produto,
preço do produto e quantidade desejada.
Calcular o valor total da compra: multiplicar o preço do produto
pela quantidade desejada.
Exibir o valor total da compra para o usuário.
(b) Adicione um desconto de 10% para compras acima de R$ 100,00.
(c) Implemente a opção de pagamento com cartão de crédito ou dinheiro.

```python
nome = input('Digite seu nome: ')
nomeProduto = input('Digite o nome do produto: ')
precoProduto = float(input('Digite o valor do produto: '))
qtidade = int(input('Digite a quantidade a ser comprada: '))

print(f'Seja bem vindo(a) {nome}, segue o resumo de sua compra: ')
valorTotal = precoProduto * qtidade
desconto = valorTotal * 10 / 100
valorTotalComDesconto = valorTotal - desconto

if valorTotal > 100:
    print(f'{nome} o valor total de sua compra é de: {valorTotal:.2f} e com desconto ficará {valorTotalComDesconto:.2f} ')
else:
    print(f'Valor total de sua compra é de {valorTotal:.2f}')

pagamento = input('Como sera a forma de pagamento?  (C) => cartao de credito  /  (D) => dinheiro: ' )
if pagamento == 'c'.lower():
    print('Pagamento efetuado com Cartão de crédito!')
else:
      print('Pagamento efetuado com Dinheiro!')
```
